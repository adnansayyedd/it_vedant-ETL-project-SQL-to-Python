```python
# ---Queies---

# 1. List all students with their enrolled course names (JOIN)
query = '''SELECT s.Student_id, s.First_name, c.Course_name
FROM Student_info s
JOIN Course_info c ON s.Student_id = c.Student_id'''
df = pd.read_sql(query, engine)
df
```

```python
# 2. Get each students course and the trainer assigned (JOIN)
query = '''SELECT s.First_name AS Student, c.Course_name, t.First_name AS Trainer
FROM Student_info s
JOIN Course_info c ON s.Student_id = c.Student_id
JOIN Trainer_info t ON c.Trainer_id = t.Trainer_id'''
df = pd.read_sql(query, engine)
df
```

```python
# 3. Show placements with student and job role details (JOIN)
query = '''SELECT s.First_name, p.Placement_companies_name, p.Placement_job_role, p.Placement_salary_package
FROM Student_info s
JOIN Placement_info p ON s.Student_id = p.Student_id'''
df = pd.read_sql(query, engine)
df
```

```python
# -- 4. Find students who are not placed yet (SUBQUERY)
query = '''SELECT * FROM Student_info
WHERE Student_id NOT IN (
    SELECT Student_id FROM Placement_info
)'''
df = pd.read_sql(query, engine)
df
```

```python
# -- 5. Get the trainer with the highest salary (SUBQUERY)
query = '''SELECT * FROM Trainer_info
WHERE Salary = (
    SELECT MAX(Salary) FROM Trainer_info
)'''
df = pd.read_sql(query, engine)
df
```

```python
# -- 6. Find students enrolled in the most expensive course (SUBQUERY)
query = '''SELECT s.First_name, c.Course_name, c.Course_fees
FROM Student_info s
JOIN Course_info c ON s.Student_id = c.Student_id
WHERE c.Course_fees = (
    SELECT MAX(Course_fees) FROM Course_info
)'''
df = pd.read_sql(query, engine)
df
```

```python
#-- 7. Count of students per course
query = '''SELECT Course_name, COUNT(Student_id) AS Total_Students
FROM Course_info
GROUP BY Course_name'''
df = pd.read_sql(query, engine)
df
```

```python
# -- 8. Number of students in each admission status
query = '''SELECT Status, COUNT(*) AS Total
FROM Admission_info
GROUP BY Status'''
df = pd.read_sql(query, engine)
df
```

```python
# -- 9. Average trainer rating per trainer
query = '''SELECT t.First_name, AVG(f.Trainer_rating) AS Avg_Rating
FROM Feedback_info f
JOIN Trainer_info t ON f.Trainer_id = t.Trainer_id
GROUP BY t.First_name'''
df = pd.read_sql(query, engine)
df
```

```python
# -- 10. List all students with 'Pending' admission status
query = '''SELECT s.First_name, a.Status
FROM Student_info s
JOIN Admission_info a ON s.Student_id = a.Student_id
WHERE a.Status = 'Pending';'''
df = pd.read_sql(query, engine)
df
```

```python
# -- 11. List courses that have more than 1 student enrolled
query = '''SELECT Course_name, COUNT(Student_id) AS Student_Count
FROM Course_info
GROUP BY Course_name
HAVING COUNT(Student_id) > 1;'''
df = pd.read_sql(query, engine)
df  
```

```python
#-- 12. Display students placed with salary above 6 LPA

query = '''SELECT s.First_name, p.Placement_companies_name, p.Placement_salary_package
FROM Student_info s
JOIN Placement_info p ON s.Student_id = p.Student_id
WHERE p.Placement_salary_package > 600000;'''

df = pd.read_sql(query, engine)
df
```

```python
# -- 13. List students with their admission date (recent first)
query = '''SELECT s.First_name, a.Admission_enrolled_date
FROM Student_info s
JOIN Admission_info a ON s.Student_id = a.Student_id
ORDER BY a.Admission_enrolled_date DESC;'''

df = pd.read_sql(query, engine)
df
```

```python
# 14. Show feedback for courses rated below 5
query = '''SELECT s.First_name, c.Course_name, f.Course_feedback
FROM Feedback_info f
JOIN Student_info s ON f.Student_id = s.Student_id
JOIN Course_info c ON f.Course_id = c.Course_id
WHERE f.Course_feedback < 5;'''
df = pd.read_sql(query, engine)
df
```

```python
# -- 15. HR issues reported by each trainer
query = '''SELECT t.First_name, h.Issues
FROM Trainer_info t
JOIN HR_info h ON t.Trainer_id = h.Trainer_id
WHERE h.Issues IS NOT NULL;'''

df = pd.read_sql(query, engine)
df
```
